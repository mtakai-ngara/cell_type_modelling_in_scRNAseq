```python
#%%script false
#fname='../../notebooks_backup/python/results/scma/mESC/dimensionality_reduction_method_comparison/quantile_0.0_1.0_preselected_quantile_none_var_genes_removed/hs_scores.tab'
#homo_scores_df=utils.read_inputdata(filepath=fname)
hs_df=pd.read_pickle(path='../results/data/dimensionality_reduction/dimensionality_scores.gz',compression='gzip')
split_groups =['modification_method', 'clustering_method', 'randomize_cells_labels',
               'top_var_genes_cut_off','top_var_genes_included','post_mod_genes','bin_start',
               'bin_end','cells_count','protocol']
replicates_grps=['cells_count','modification_method', 'clustering_method','fc','genes_count', 
                 'randomize_cells_labels','method','top_var_genes_cut_off','top_var_genes_included',
                 'normalization_method', 'pca_components','bin_end','protocol','cells_count']
replicates_cut_off=5
out_dir=None
out_dir='../results/dimensionality_reduction/figures'
fc_list=[0.2,0.3,3,5]   
#fc_list=None
#fc_order=[0.2,0.3,3,5]
fc_order=None
x_axis_subset=[5,10,30,50]
#x_axis_subset=None
in_col_wrap=2
show_reps_counts=['fc','genes_count','method']
show_reps_counts=None
hue_order=None
#Globala_variables
sns.set_context(context='paper',rc={'style':'white'})
#sns.set_palette(palette='colorblind')
#palette_dict='bright'
hue='method'
x_axis='genes_count'
limit=100
temp_palette=sns.color_palette(palette='tab10')
boxprops_dict=dict(linewidth=0)
boxplot_mean=True
in_kind='bar' # 'bar','box', 'point'
fig=utils.plot_classification_scores_between_experiments(expriment_path=hs_df, out_dir=out_dir,
                                                         x_axis=x_axis,split_groups=split_groups,
                                                         replicates_grps=replicates_grps,
                                                         y_axis='adj_rand_index',in_kind=in_kind,ci=95,
                                                         in_col_wrap=in_col_wrap,
                                                         replicates_per_grp=replicates_cut_off,
                                                         col_subset=fc_list,hue=hue,hue_order=hue_order,
                                                         x_axis_subset=x_axis_subset,col_order_value=fc_order,
                                                         show_reps_counts=show_reps_counts,subplot_title_size=10,
                                                         palette=temp_palette,facet_size=5,add_swarm=False,
                                                         swarm_points_size=5.0,x_axis_rotation=70,
                                                         boxplot_mean=boxplot_mean,boxplot_outliers=False,
                                                         boxprops=boxprops_dict,limit=limit,
                                                         fig_title='Dimensionality reduction') 
```
