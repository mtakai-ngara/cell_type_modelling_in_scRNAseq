```python
import pandas as pd
import celldiscoveryutilities as utils
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#%%script false
fname='../results/data/clustering_method/clustering_method_scores.gz'
split_groups = ['method', 'modification_method','randomize_cells_labels','top_var_genes_cut_off','cells_count']
replicates_grps=['method', 'cells_count', 'genes_count', 'fc', 'modification_method','clustering_method',
                 'randomize_cells_labels']
homo_scores_df=pd.read_pickle(path=fname,compression='gzip')
out_dir=None
out_dir='../results/clustering_methods/figures'
hue='clustering_method'
replicates_grps=None
replicates_cut_off=90
#fc_list=[0.0,0.2,0.3,3]     
fc_list=None
fc_order=[0.2,0.3,2,5]
#fc_order=None
x_axis_subset=[10,30,50]
#x_axis_subset=None
in_col_wrap=2
show_reps_counts=['fc','genes_count',hue]
show_reps_counts=None
hue_order=None
#Globala_variables
sns.set_context(context='paper',rc={'style':'white'})
#sns.set_palette(palette='colorblind')
#palette_dict='bright'
x_axis='genes_count'
temp_palette=sns.color_palette(palette='Set1')
boxprops_dict=dict(linewidth=0)
boxplot_mean=False
in_kind='box'
limit=100
temp_palette={'knn': 'purple','kmeans': 'b','hdbscan': 'g'}
fig=utils.plot_classification_scores_between_experiments(expriment_path=homo_scores_df, out_dir=out_dir,
                                                         x_axis=x_axis,split_groups=split_groups,
                                                         replicates_grps=replicates_grps,
                                                         y_axis='adj_rand_index',in_kind=in_kind,ci=95,
                                                         in_col_wrap=in_col_wrap,
                                                         replicates_per_grp=replicates_cut_off,
                                                         col_subset=fc_list,hue=hue,hue_order=hue_order,
                                                         x_axis_subset=x_axis_subset,col_order_value=fc_order,
                                                         show_reps_counts=show_reps_counts,
                                                         palette=temp_palette,facet_size=5,subplot_title_size=10,
                                                         add_swarm=False,
                                                         swarm_points_size=5.0,x_axis_rotation=70,
                                                         boxplot_mean=boxplot_mean,boxplot_outliers=False,
                                                         boxprops=boxprops_dict,limit=limit,fig_title='Clustering methods comparison') 
```
